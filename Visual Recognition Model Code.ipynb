```python
#Adding a training object (called "car") to an example image
import json
from ibm_watson import VisualRecognitionV4
from ibm_watson.visual_recognition_v4 import TrainingDataObject, Location
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator

authenticator = IAMAuthenticator('bfIt1-So8CljKR5b-ZYoBAtLcXCb4vltq8Q3NrCHwlCQ')
visual_recognition = VisualRecognitionV4(
    version='2019-02-11',
    authenticator=authenticator
)

visual_recognition.set_service_url('https://api.us-south.visual-recognition.watson.cloud.ibm.com/instances/1cbd8c01-6315-4d78-b1de-2bc83f870129')

result = visual_recognition.add_image_training_data(
    collection_id='d7e670d3-e6a6-42df-b14f-5b59d4f7bf80',
    image_id='0846bb29-b97d-444e-8366-1ad21a1_656fcd1e64f422faf60d4e8949882c6e',
    objects=[
        TrainingDataObject(object='car',
                           location=Location(5, 13, 760, 419))
    ]).get_result()
print(json.dumps(result, indent=2))
```
