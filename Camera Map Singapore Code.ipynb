```python
# instantiate a feature group for the cameras in the dataframe
cameras = folium.map.FeatureGroup()

# loop through the 100 cameras and add each to the incidents feature group
for lat, lng, label in zip(df_data_1.location_latitude, df_data_1.location_longitude, df_data_1.location):
    cameras.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='black',
            fill=True,
            popup = label,
            fill_color='white',
            fill_opacity=1
        )
    )

# add incidents to map
my_map.add_child(cameras)
```
